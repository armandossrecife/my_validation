```python
import pandas as pd
import random

def create_my_zip(filename, my_list):
  my_issues = ""
  my_file = filename
  for issue in my_list:
    my_issues = my_issues + ' '+ issue

  my_issues = ' '.join(my_list)
  my_issues = my_issues.replace('CASSANDRA', '/content/my_issues/CASSANDRA')
  my_command = 'zip' + ' ' + my_file + ' ' + my_issues
  return my_command

def create_my_df(my_list):
  my_dict = {}
  list_issue_id, list_type, list_summary = list(), list(), list()
  for each in my_list:
    filename = '/content/my_issues/' + each
    with open(filename, mode='r') as f_temp:
      first_line = f_temp.readline()
      first_line = first_line.split(':')
      first_line = first_line[-1].strip()
      second_line = f_temp.readline()
      second_line = second_line.split(':')
      second_line = second_line[-1].strip()
      list_issue_id.append(each)
      list_type.append(first_line)
      list_summary.append(second_line)
  my_dict = {
          'issue_id': list_issue_id,
          'issue_type': list_type,
          'summary': list_summary
      }
  my_df = pd.DataFrame(my_dict)
  return my_df
```

```python
list_isues_temp = !ls -l my_issues
del list_isues_temp[0]
list_issues = []
for issue in list_isues_temp:
  issue_temp = issue.split(' ')
  list_issues.append(issue_temp[-1])

print(f'Lista de {len(list_issues)} issues inspecionados: {list_issues}')
```

```python
lista_armando = random.sample(list_issues, 23)
lista_ricardo = random.sample(list_issues, 23)
lista_lincoln = random.sample(list_issues, 23)

print('Seleciona 23 issues aleatoriamente para cada participante')
print(f'Lista de {len(lista_armando)} issues Armando: {lista_armando}')
print(f'Lista de {len(lista_ricardo)} issues Armando: {lista_ricardo}')
print(f'Lista de {len(lista_lincoln)}issues Armando: {lista_lincoln}')
```

```python
df_armando = create_my_df(my_list=lista_armando)
df_ricardo = create_my_df(my_list=lista_ricardo)
df_lincoln = create_my_df(my_list=lista_lincoln)
```
